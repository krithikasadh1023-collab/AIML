```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = {
    'Annual_Income': [15, 16, 17, 18, 80, 82, 85, 90, 25, 30],
    'Spending_Score': [39, 40, 41, 42, 70, 72, 75, 78, 50, 55]
}
```

```python
df = pd.DataFrame(data)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)
```

```python
score = silhouette_score(df[['Annual_Income', 'Spending_Score']], df['Cluster'])
print(df)
print("Silhouette Score:", score)
```

```python
plt.scatter(df['Annual_Income'], df['Spending_Score'], c=df['Cluster'])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation using K-Means")
plt.show()
```
